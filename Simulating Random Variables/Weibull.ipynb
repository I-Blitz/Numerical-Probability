### Q1   
Soit X une v.a.r suivant une loi de Weibull de paramètre $k,\lambda,\theta$ avec $k > 0$, $\lambda > 0$   
On sait que $f_X(x) = \frac{k}{\lambda} \Bigl(\frac{x-\theta}{\lambda} \Bigr)^{k-1} \exp \Bigl( - (\frac{x-\theta}{\lambda})^{k} \Bigr) \mathbb{1}_{ \{ x \geq 0 \} } $    
On cherche à définir la fonction de répartition de X.    
Tout d'abord on remarque que si $x < 0$, $\mathbb{P} \Bigl( X \leq x \Bigr) = 0$   
Soit $x \in \mathbb{R}^{+}$     
$$
\begin{align*}
    \mathbb{P} \Bigl( X \leq x \Bigr) &= \int_{- \infty}^{x} f_{X}(u) \mathrm{d}u  \\
    &= \int_{0}^{x} \frac{k}{\lambda} \Bigl(\frac{u-\theta}{\lambda} \Bigr)^{k-1} \exp \Bigl( - (\frac{u-\theta}{\lambda})^{k} \Bigr)   \mathrm{d}u \\
    &= \Bigl[- \exp \Bigl( - (\frac{u-\theta}{\lambda})^{k} \Bigr) \Bigr]_{0}^{x}  \\
    &= 1 - \exp \Bigl( - (\frac{x-\theta}{\lambda})^{k} \Bigr)
\end{align*}
$$

Donc on a la fonction de répartition $F_X$ suivante :   
$$
\begin{equation*}
    F_{X}(x) =  \Bigl( 1 - \exp \bigl( - (\frac{x-\theta}{\lambda})^{k} \bigr)  \Bigr) \mathrm{1}_{ \{ x \geq 0 \} } 
\end{equation*}
$$

Maintenant on supposera que $\theta = 0$

### Q2 
On souhaite déterminer $\mathbb{E} \Bigl[ X \Bigr] $ et $ \mathbb{V} \Bigl[ X \Bigr] $   
$$
\begin{align*}
    \mathbb{E} \Bigl[ X \Bigr] &= \int_{-\infty}^{+ \infty} x f_{X}(x) \mathrm{d}x &\quad \text{théorème de transfert } \\
    &= \int_{0}^{+ \infty} \frac{k}{\lambda} x \Bigl(\frac{x}{\lambda} \Bigr)^{k-1} \exp \Bigl( - (\frac{x}{\lambda})^{k} \Bigr)   \mathrm{d}x \\
\end{align*}
$$
On pose le changement de variable suivant $t = \frac{x}{\lambda}$ qui est légitime car affine (donc $C^{1}$, bijectif, monotone).
$$ 
\begin{align*} 
    \mathbb{E} \Bigl[ X \Bigr] &= \lambda k \int_{0}^{+ \infty} t^{k} \exp \Bigl( - t^{k} \Bigr)   \mathrm{d}t \\
    &= \lambda \int_{0}^{+ \infty} u \exp( -u) u^{\frac{1}{k} -1} \mathrm{d}u &\quad \text{changement de variable :} u = t^{k} \text{ légitime car } C^{1} \text{bijectif de } R^{+} \text{ dans } R^{+}  \\
    &= \lambda \int_{0}^{+ \infty} u^{\frac{1}{k}} \exp(-u) \mathrm{d}u \\
    &= \lambda \Gamma \bigl( 1 + \frac{1}{k} \bigr) \\
\end{align*}
$$
$$
\begin{align*}
    \mathbb{V} \Bigl[ X \Bigr] &= \mathbb{E} \Bigl[ X^{2} \Bigr] - \mathbb{E} \Bigl[ X \Bigr]^{2} \\
\end{align*}
$$
$$ 
\begin{align*} 
    \mathbb{E} \Bigl[ X^{2} \Bigr] &= \int_{- \infty}^{+ \infty} x^{2} f_{X}(x) \mathrm{d}x &\quad \text{théorème de transfert} \\
    &= \int_{0}^{+ \infty} x^{2} \frac{k}{\lambda} \Bigl(\frac{x}{\lambda} \Bigr)^{k-1} \exp \Bigl( - (\frac{x}{\lambda})^{k} \Bigr)   \mathrm{d}x  \\
    &= \lambda^{2} k \int_{0}^{+ \infty} u^{k+1} exp(-u^{k}) \mathrm{d}u &\quad \text{changement de variable : u = } \frac{x}{\lambda}   \\
    &= \lambda^{2} k \int_{0}^{+ \infty} t^{1+\frac{1}{k}} exp(-t) \frac{1}{k} t^{\frac{1}{k} -1} \mathrm{d}t &\quad \text{changement de variable : } t = u^{k} \text{ légitime car } C^{1} \text{bijectif de } R^{+} \text{ dans } R^{+}  \\
    &= \lambda^{2} \int_{0}^{= \infty} t^{\frac{2}{k}} exp(-t) \mathrm{d}t 
    &= \lambda^{2} \Gamma \bigl( 1 + \frac{2}{k} \bigr) 
\end{align*}
$$
D'où le résultat suivant : 
$$
\begin{equation*}
    \mathbb{V} \Bigl[ X \Bigr] = \lambda^{2} \Gamma \bigl( 1 + \frac{2}{k} \bigr) - \lambda \Gamma \bigl( 1 + \frac{1}{k} \bigr)
\end{equation*} 
$$

### Q3.a)
Posons $Y = F_{X}(X)$, on remarque que Y est à valeurs dans $[0,1]$.
Soit $u \in [0,1]$
$$
\begin{align*}
    \mathbb{P}(Y \leq u) &= \mathbb{P}( F_{X}(X) \leq u)\\
    &= \mathbb{P} ( X \leq F_{X}^{-1}(u) ) &\quad \text{légitime car } F_{X}(X) \text{ est continue strictement croissante} \\
    &= F_{X}( X \leq F_{X}^{-1}(u) ) \\
    &= u 
\end{align*}
$$
   
   
Donc $Y \sim \mathbb{U}_{[0,1]} $ Donc si on peut inverser $F_{X}$, on peut à partir d'un tirage uniforme sur $[0,1]$ générer un tirage de X.   
Soit $u \in [0,1]$ 
$$
\begin{align*}
    u &= F_{X}(x) \\
    u &= 1 - exp(-(\frac{x}{\lambda}^{k})) \\
    exp(-(\frac{x}{\lambda}^{k})) &= 1 - u \\
    -\bigl( \frac{x}{\lambda} \bigr)^{k} &= ln(1-u) \\
    \bigl(\frac{x}{\lambda} \bigr)^{k} &= -ln(1-u) \\
    \frac{x}{\lambda} &= \bigl( -ln(1-u) \bigr)^{\frac{1}{k}} \\
    x &= \lambda \bigl( -ln(1-u) \bigr)^{\frac{1}{k}} \\
\end{align*}
$$

```python
N = 10000
theta = 0
lmbd = 2
k = 3
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rweibull(N = N, theta = theta, lmbd = lmbd, k = k):
    res = np.zeros(N)
    U = np.random.rand(N)
    for i in range(N):
        res[i] = lmbd *  ( (-np.log(1-U[i]) ) ** (1/k) )
    return res

WeibullDensity = lambda theta,lmbd,k,x : k/lmbd * (  ((x-theta)/lmbd)**(k-1) ) * np.exp(-( (x-theta)/lmbd )**k ) * (x >= theta)

    
```

Avant de passer à la suite, nous allons vérifier que ce que l'on a programmé est cohérent avec ce que l'on devrait avoir. On va donc faire un histogramme de tirages et comparer cela a la densité théorique

```python
W = rweibull()
val = np.linspace(np.amin(W),np.amax(W),N)
plt.subplots(figsize=(10,6))
plt.hist(W, density = True, alpha = 0.5, bins = 20, label = "tirage selon Weibull ")
plt.plot(val,WeibullDensity(theta,lmbd,k,val), color = "r", label = "densité théorique")
plt.title("tirage de Weibull depuis uniforme")
plt.legend(loc = 1)
plt.show()
```

```python
vec_n = [30,100,1000,10000]
x = np.linspace(0,5,500)

fctRepart = lambda theta,lmbd,k,x : 1 - np.exp(- ( (x-theta)/lmbd )**k ) * (x >= theta)

def RepartitionEmpirique(x,W):
    res = np.zeros(len(x))
    for i in range(len(x)):
        for j in range(len(W)):
            res[i] += (W[j] < x[i])
    res /= len(W)
    return res

plt.subplots(figsize=(9,6))
for i in range(len(vec_n)):    
    W = rweibull(N = vec_n[i])
    Repartition = RepartitionEmpirique(x,W)
    plt.plot(x,Repartition, label = "n = {}".format(vec_n[i]) )
    
plt.plot(x,fctRepart(theta,lmbd,k,x), color = "r", label = "fonction de répartition théorique")
plt.title("fonction de répartition de Weibull")
plt.legend(loc = 4)
plt.show()
```

#### Commenter le graphique 
On remarque que plus le nombre de tirages de notre loi de Weibull est grand, plus notre fonction de répartition empirique se rapproche de la fonction de répartition théorique. Ce qui est assez cohérent car finalement notre approche empirique est basée sur Monte-Carlo et on a simplement dit que $\forall u \in \mathbb{R}^{+}$, $ \mathbb{P} \Bigl( X \leq u \Bigr) = \mathbb{E} \Bigl[ \mathbb{1}_{ \{ X \leq u \} } \Bigr] $

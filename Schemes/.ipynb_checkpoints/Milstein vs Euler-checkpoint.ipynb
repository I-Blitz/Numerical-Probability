# Milstein vs Euler 

comparaison numérique / analytique
$$
\begin{equation*}
    S_{t} = s_{0} \exp{(\sigma W_{t} + (r- \frac{\sigma^2}{2})t)} \\
     S_{0} = s_{0}
\end{equation*}
$$

```python
## solution analytique
prixActif = lambda t,W,sigma,r, s0: s0*np.exp(sigma*W + t*(r -  (sigma**2)/2 )) 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time
```

```python
### paramètres "physique"
T = 1
r = 0.1
S0 = 100
sigma = 5

### paramètres "numériques"
N = 100*T
h = T/N
```

```python
## Ici on construit la matrice de variance-covariance du mouvement brownien
def MatBrownien(t,N):
    Mat = np.zeros( (N+1,N+1) )
    for i in range(1,N+1):
        for j in range(1,N+1):
            Mat[i][j] = min(t[i],t[j])
    return Mat 

## Ici on crée le vecteur W qui est censé représenter un mouvement brownien sur l'intervalle [0,T]
def MvmtBrownien(T,N):
    h = T/N
    W = np.random.normal(0,1, size = N+1)
    t = np.array([i*h for i in range(N+1)])
    Mat = MatBrownien(t,N)
    A = np.linalg.cholesky(Mat[1:,1:])
    B = np.zeros((N+1,N+1))
    B[1:,1:] = A
    W = np.dot(B,W)
    return (t,W)
```

```python
def SDE_comparaison(S0,r,sigma,T,N):
    t,W = MvmtBrownien(T,N)
    S_euler = np.zeros(N+1)
    S_milstein = np.zeros(N+1)

    S_euler[0] = S0
    S_milstein[0] = S0
    S_analytique = prixActif(t,W,sigma,r,S0)

    for i in range(N):
        S_euler[i+1] = S_euler[i] + r*S_euler[i]*h + sigma*S_euler[i]*(W[i+1]-W[i])
        S_milstein[i+1] = S_milstein[i] + r*S_milstein[i]*h + sigma*S_milstein[i]*(W[i+1]-W[i]) + 1/2 * sigma * S_milstein[i] * sigma *( (W[i+1]-W[i])**2 - h )
        
    return t,S_euler,S_milstein,S_analytique
```

```python
sigmaVal = np.linspace(0.5,5,10)
for sigma in sigmaVal :
    t,S_euler,S_milstein,S_analytique = SDE_comparaison(S0,r,sigma,T,N)
    plt.subplots(figsize=(8,6))
    plt.plot(t,S_euler, label = "euler", color = 'g')
    plt.plot(t,S_milstein, label = "milstein", color = 'b')
    plt.plot(t,S_analytique, label = "sol analytique", color = 'r')
    plt.xlabel("temps")
    plt.ylabel("valeurs")
    plt.title("comparaison numérique, analytique avec $\sigma$ = {}".format(sigma))
    plt.legend()
    plt.plot()
```

Plus la vol $\sigma$ est élevée, plus on voit les limites du schéma d'Euler. Notre solution approchée par Euler s'éloigne très fortement de la réalité et on a même un cas où notre solution devient négative ce qui est impossible.
